# Exercice 2

```python
r= float(input("Entrez le rayon : "))
h= float(input("Entrez la hauteur : "))

def volume_cone(rayon = float, hauteur= float):
    from math import pi
    return (1/3)* pi * rayon**2 * hauteur
     
print(f"Le volume du cône est: {volume_cone(r, h)}")
```

# Exercice 5

```python
def pair(nombre_entier =int)-> str:
    if nombre_entier <0 :
        return "NOT POSITIVE"
    if nombre_entier % 2 == 0:
        return "PAIR"
    else:
        return "IMPAIR"

while True:
    n= int(input("Entrez un nombre entier positif: "))
    resultat= pair(n)
    if resultat != "NOT POSITIVE":
        print(f"Le nombre est {pair(n)}")   
    else : 
        print ("Le nombre entré est négatif")
        break




    



```

# Exercice 7

```python
def plus_ou_moins(min=int,max=int,tries=int):
    from random import randint  
    nombre_secret = randint(min, max)
    nombre_proposé = int(input(f"Devinez le nombre secret entre {min} et {max}], vous avez {tries} essais : "))
    for i in range(tries):
        if nombre_proposé < nombre_secret:  
            print(f"Trop bas ! Essayez encore")
        elif nombre_proposé > nombre_secret:   
            print(f"Trop haut ! Essayez encore")
        else:
            print(f"Félicitations ! Vous avez deviné le nombre secret en {i+1} essais.")
            break
    
        nombre_proposé = int(input(f"Devinez le nombre secret entre {min} et {max} : "))
        
    else:
        print(f"Désolé, vous avez utilisé tous vos essais. Le nombre secret était {nombre_secret}.")

plus_ou_moins(0,100,7)
```

```python
def notes():
    nombre_notes=0
    somme_notes=0
    liste_notes=[]
    note= (input("Entrez une note entre 0 et 20, n'entrez rien pour terminer : "))
    
    while note != "":
        note_float= float(note)
        if 0 <= note_float <= 20:
            liste_notes.append(note_float)
            somme_notes += note_float
            nombre_notes += 1
        else:
            print("Note invalide, veuillez entrer une note entre 0 et 20.")
        note= (input("Entrez une note entre 0 et 20, n'entrez rien pour terminer : "))
    
    else:
        if nombre_notes > 0:
            moyenne = somme_notes / nombre_notes
            return [nombre_notes,max(liste_notes),min(liste_notes),moyenne]
        else:
            return 

resultat_final = notes()
if resultat_final == None:
    print (f"Aucune note entrée")
else:
    print (resultat_final)
```

# Exercice 10

```python
s=str(input("Entrez une chaîne de caractères"))
def astérisques(s):
    chaîne_astérisque = ""
    liste_char = []
    for char in s:
        liste_char.append(char)
    
    for i in range(len(liste_char)):
        chaîne_astérisque += liste_char[i] + "*"

    return chaîne_astérisque    
    

    

print (astérisques(s))

```
