# Exercice 2

```python
r= float(input("Entrez le rayon : "))
h= float(input("Entrez la hauteur : "))


def volume_cone(r, h):
    from math import pi
    v = (1/3) * pi * r**2 * h
    print(f"Le volume du cône est: {v}")


volume_cone(r, h)
```

# Exercice 3

```python
p= float(input("Entrez votre poids en kg : "))
t= float(input("Entrez votre taille en mètre : "))

def imc(p, t):
    imc_value = p / (t ** 2)
    print(f"Votre IMC est: {imc_value}")


imc(p,t)
```

# Exercice 6

```python
def révision_multiplication():
   score = 0
   for i in range(1,6):
        
        from random import randint
        a= randint(1,10)
        b= randint(1,10)
        réponse=int(input(f"Combien font {a} x {b} ? "))
        if réponse== a*b:
            print(f"Correct! Vous gagnez un point.")
            score += 1
        else:
            print(f"Incorrect. La bonne réponse est {a*b}, vous perdez un point.")       
            score -= 1
   print(f"Votre score final est {score} points.")
         
révision_multiplication()

      
```
