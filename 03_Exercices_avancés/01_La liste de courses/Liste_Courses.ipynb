# **La Liste de course**

```python
import json
import os

fichier_json = "ma_liste.json"

if os.path.exists(fichier_json):                #si le fichier json existe déja on charge la liste qu'il contient
    with open(fichier_json, "r") as f:
        liste_élement = json.load(f)
else :
    liste_élement = []                          #sinon on part d'une liste vide

affichage = """
Choisissez une option:
\t1: Ajouter un élément
\t2: Enlever un élément
\t3: Afficher la liste
\t4: Vider la liste
\t5: Sauvegarder la liste
"""

def entrer_choix():
    choix = input("Entrez votre choix (1-5)")                   #la fonction qui vérifie le choix de l'utilisateur
    while choix not in ["1", "2", "3", "4", "5"]:
        choix = input("Choix invalide, Entrez votre choix (1-5)")
    return choix
print(affichage)
while True:
    choice = entrer_choix()                                                 #boucle qui réalise une action à chaque entrée de l'utilisateur et lui demande une nouvelle entrée
    match choice:
        case "1":
            liste_élement.append(f"élément{len(liste_élement)+1}")
            print (f"Elément ajouté")
        case "2":
            try:
                liste_élement.remove(liste_élement[-1])
            except IndexError:
                continue
            print (f"Le dernier élément a été retiré")
        case "3":
            print (f"Votre liste est la suivante pour l'instant : {liste_élement}")
        case "4":
            liste_élement.clear()
            print (f"Liste vidée")
        case "5":
            with open(fichier_json, "w") as f:                                  #Sauvegarde du panier dans le fichier JSON adéquat, création si il n'existe pas. Sortie de la boucle
                json.dump(liste_élement, f)
            print("Liste sauvegardée dans le fichier JSON.\n")
            break
        case _:
            print("choix invalide")
            break
   
```
