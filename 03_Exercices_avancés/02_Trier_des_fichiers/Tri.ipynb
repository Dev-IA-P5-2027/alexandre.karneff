**Trier_des_fichiers**

```python
import os
import shutil

file = 'Fichiers/example.mp3'


def Tri_Fichier(file_path):
    try:
        file_name, file_extension = os.path.splitext(file_path)
        file_extension = file_extension.lower()

        print("File Name:", file_name)
        print("File Extension:", file_extension)

        destinations = {
            ('.jpg', '.png'): 'Images',
            ('.pdf', '.doc', '.docx', '.txt'): 'Documents',
            ('.mp3'): 'Musiques',
            ('.mp4','.mkv'): 'Videos'
            }

        destination = None
        for extensions, folder in destinations.items() :
            if file_extension in extensions:
                destination = folder
                break
        
        if destination:
            destination_directory = os.path.join('', destination)               #Cr√©er le dossier si il n'existe pas
            os.makedirs(destination_directory, exist_ok=True)

            destination_path = os.path.join(destination_directory, os.path.basename(file_path))         #Copier le fichier dans le dossier destination 
            shutil.copy(file_path, destination_path)
            print(f"File sent to {destination}")
        
        else :
            print("Extension Invalide")

    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found")


Tri_Fichier(file)
```
