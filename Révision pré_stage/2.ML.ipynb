1️⃣ **Qu’est-ce que l’apprentissage supervisé et non supervisé ?**  

```python
# Réponse :
```

On parle d'apprentissage supervisé lorsqu'on a les données de notre variable cible (target) pour l'entrainement de notre modèle, on peut donc l'entrainer en prenant en compte ces valeurs pour l'évaluer/les métriques. Associer des entrées à des sorties connues<br>
On parle d'apprentissage non supervisé lorsque l'on a pas les données de notre variable cible (target) pour l'entrainement de notre modèle, on doit donc l'entrainer et l'évaluer différemment, le modèle cherche à trouver des pattern cf clustering

3️⃣ **Qu'est-ce qu'un algorithme de clustering ?**  

```python
# Réponse :
```

Un algorithme de clustering (ex: k-means, k-medoids) est un algorithme de ML qui sert dans le cas de l'apprentissage non-supervisé, il regroupe les données sans avoir de target donc, selon les similarités entre celles-ci (par exemple les distances), dans le but de créer des "cluster" sensés regrouper une classe ou des données plus proches les unes des autres etc.

4️⃣ **Qu’est-ce qu’un biais et une variance dans un modèle de Machine Learning ?**  

```python
# Réponse :
```

Le biais mesure l'écart entre la valeur prédite par un modèle et la valeur réelle, c'est l'erreur du modèle qui n'arrive pas à capturer les pattern de données. Approximation excessive du modèle (underfitting) <br>
La variance est elle aussi une erreur représentant la les variations du modèle face aux fluctuations des données d'entrainement, lorsque le modèle est trop entrainé sur le bruit du training set. Sensibilité excessive aux données d'entrainement, entrainement en surajustement (overfitting)

5️⃣ **Pourquoi et comment diviser un dataset en train/test ?**  

```python
# Réponse :
```

On doit diviser notre dataset en un training set un test set, notre training est passé au modèle pour son entrainement (on prend en général 80% du set complet pour le training),
le test set lui sert à évaluer les performances du modèle après son entrainement (en général les 20% restant du set). <br>
Il est important de diviser notre set pour que le modèle s'entraine avec certaines données, mais soit évalué sur des données qu'il n'a "jamais vu", c'est à dire dont il ne s'est pas servi pour l'entrainement. <br>
Cela est nécessaire pour évaluer la capacité de notre modèle à généraliser à de nouvelles données, et de ne pas juste mémoriser les pattern et bruits de l'entrainement
<br>
On split avec la méthode train_test_split de scikit-learn en général

6️⃣ **Comment choisir un bon algorithme de Machine Learning ?**  

```python
# Réponse :
```

Pour choisir un bon algorithme de Machine Learning il faut identifier le problème auquel nous sommes confronté dans un premier temps : Classification, Régression, Clustering ?
<br>
Ensuite, il y'a des indicateurs qui peuvent donner des indices sur le type d'algorithme a utiliser, par exemple : beaucoup de données catégorielles, numériques, données linéaires, nombre de données, de features que l'on a choisi etc. ou alors se servir de son expérience
<br>
On fait des tests avec plusieurs algorithmes qui semblent pertinents (ceux pré-choisis ou plusieurs types différents si on ne sait pas) et on les évalue tous rapidement dans un premier temps.
<br>
En fonction des métriques qui nous intéressent on peut déja détacher un ou plusieurs algorithmes qui peuvent sembler intéressant à optimiser au vu de ces premiers résultats.
Il faudra ensuite optimiser les hyper paramètres pour essayer d'en améliorer les performances et choisir le meilleur

7️⃣ **Quels sont les principaux indicateurs de performance d’un modèle (Accuracy, MSE, F1-score, etc.) ?**  

```python
# Réponse :
```

Les métriques dépendent du type d'algorithme de ML que l'on utilise :
Accuracy, Precision, Recall, F1-score ...: Classification <br>
MSE, MAE, RMSE, R2 ... : Régression <br>
Rand index, ARI, AMI, Silhouette coefficient ... : Clustering <br>

Accuracy : Pourcentage de prédictions correctes <br>
MSE : moyenne des carrés de l'erreur<br>
F1-score : moyenne harmonique entre précision et recall
