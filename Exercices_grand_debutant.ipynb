# Exercice 13

```python
a= int(input("Entrez le premier nombre entier: "))
b= int(input("Entrez le deuxième nombre entier: "))
while b == 0:
    print("Erreur : division par zéro impossible.")
    b= int(input("Entrez un autre nombre entier différent de zéro: "))

quotient = a // b
print(f"Le quotient de {a} divisé par {b} est {quotient}")
reste = a % b
print(f"Le reste de la division de {a} par {b} est {reste}")
```

# Exercice 17


```python
s = str(input("Entrez un mot: "))
already_counted = ""
for char in s:
    count = s.count(char)
    if char in already_counted:
        continue   
    else :
        print(f"Le caractère '{char}' apparaît {count} fois dans a chaîne ")
    
    already_counted = already_counted + char





```

# Exercice 18

```python
s= str(input("Entrez une chaîne de caractère: "))
liste = []
for char in s:
    liste.append(char)

for i in liste:
    if i in 'aA':
        position = liste.index(i)
        print("Lettre 'a' trouvée à la position", position)
        for j in range(position + 1, len(liste)):
            if liste[j] in 'aA':
                position = j
                print("Lettre 'a' trouvée à la position", position)
        break
        
                   
else :
    print("Lettre 'a' non trouvée")


        
        

```

# Exercice 23

```python
s= str(input("Entrez un nom de fichier: "))
extension = ""
for i in s:
    if i == '.':
        for j in range(s.index(i)+1, len(s)):
            extension += s[j]   

print(f"L'extension du fichier est: {extension}")
```
