# Exercice 4

```python
s=str(input("Entrez une chaîne de caractères"))

def compterMots(s):
    import re
    dict_words ={}
    words=re.split(r'[;,.,(,), \s]+', s)
    for i in range (len(words)):
        word=words[i]
        if word not in dict_words: 
            dict_words[word]=1
        else:
            dict_words[word]+=1
    print (dict_words)

compterMots(s)
```

# Exercice 5

```python
def minMaxMoy(l):
    somme_nombre=0
    minimum = min(l)
    maximum = max(l)
    for nombre in l :
        somme_nombre +=nombre

    moyenne = somme_nombre/(len(l))
    # print(f"La note la moins élevée est : {minimum}")
    # print(f"La note la plus élevée est : {maximum}")
    # print(f"La moyenne est : {moyenne}")
    tuple = (minimum,maximum,moyenne)
    return(tuple)


  
print(minMaxMoy([10,18,14,20,12,16]))
```
