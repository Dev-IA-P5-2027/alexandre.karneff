# Exercice 4

```python
s=str(input("Entrez une chaîne de caractères"))

def compterMots(chaîne = str)-> dict :
    import re
    dict_words ={}
    words=re.split(r'[;,.,(,), \s]+', chaîne)
    for i in range (len(words)):
        word=words[i]
        if word not in dict_words: 
            dict_words[word]=1
        else:
            dict_words[word]+=1
    return dict_words

print(compterMots(s))
```

# Exercice 5

```python
def minMaxMoy(l=list)-> tuple:
    somme_nombre=0
    minimum = min(l)
    maximum = max(l)
    for nombre in l :
        somme_nombre +=nombre
    moyenne = somme_nombre/(len(l))
    return(minimum,maximum,moyenne)


print(minMaxMoy([10,18,14,20,12,16]))
```
