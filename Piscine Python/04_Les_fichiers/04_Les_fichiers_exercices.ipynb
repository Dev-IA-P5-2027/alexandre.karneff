```python
import os

file_path = input("Entrez le nom du fichier")
content = input ("Entrez le contenu du fichier")

def create_text_file(file, content):

    destination = 'C:/Users/alexandre.karneff/Documents/txt_test_piscine'
    file = file.lower()
    if not file.endswith(".txt"):
        file += ".txt"

    file_path = os.path.join(destination, file)
    with open (file_path, 'w') as file :
        file.write(content)

    print(f"Fichier crée avec succès dans le dossier")


create_text_file(file_path, content)
```

```python
import os

file_path = input("Quel est le nom du fichier où tripler les espaces ?")

def read_text_file(file_path):

    try :
        directory = "C:/Users/alexandre.karneff/Documents/txt_test_piscine"
        file_path = file_path.lower()
        new_file_path = file_path + "_new"

        if not file_path.endswith(".txt"):
            file_path += ".txt"
        if not new_file_path.endswith(".txt"):
            new_file_path += ".txt"
        file_path = os.path.join(directory, file_path)
        with open (file_path, 'r') as file :
            content = file.read()
            words = content.split()
            print (content, words)

        new_file_path = os.path.join(directory, new_file_path)
        with open (new_file_path, 'w') as new_file :
            result = "   "
            for word in words:
                new_content = word + "   "
                result += new_content
            new_file.write(result)

        
    except FileNotFoundError:
        print (f"Fichier {file_path} introuvable dans le directory")
    
    


read_text_file(file_path)
```
