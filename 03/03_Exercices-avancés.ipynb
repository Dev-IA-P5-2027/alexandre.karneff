# **La Liste de course**

```python
import mon_module


liste_élement = []
print("=== Choisissez une option: ===")
print("1. Ajouter un élément")
print("2. Enlever un élément")
print("3. Afficher la liste")
print("4. Vider la liste")
print("5. Sauvegarder la liste")

def entrer_choix():
    choix = input("Entrez votre choix (1-5)")
    while choix not in ["1", "2", "3", "4", "5"]:
        choix = input("Choix invalide, Entrez votre choix (1-5)")
    return choix

while True:
    choice = entrer_choix()
    match choice:
        case "1":
            liste_élement.append(f"élément{len(liste_élement)+1}")
            print (f"Elément ajouté")
        case "2":
            try:
                liste_élement.remove(liste_élement[-1])
            except IndexError:
                continue
            print (f"Le dernier élément a été retiré")
        case "3":
            print (f"Votre liste est la suivante pour l'instant : {liste_élement}")
        case "4":
            liste_élement= []
            print (f"Liste vidée")
        case "5":
            print ("Liste Sauvegardée")
            break
        case _:
            print("choix invalide")
            break
   
```
